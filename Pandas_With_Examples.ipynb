#### 6. How many person we have whose name starts with "s"?

```python
count = 0
for name in df['Name']:
    if name[0].lower() == 's':
        count += 1
print("There are {} people whose names are started with 's'".format(count))  
```

```python
# Otherway
len(df[df['Name'].str.startswith('S')==True])
```
