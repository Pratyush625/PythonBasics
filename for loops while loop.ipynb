```python
a = 1 
while a < 5 :
    print(a)
    if a == 4 :
        break
    a = a +1 
```

```python
# Infinite loop
a=1
while a<5:
    print(a)
    if a==3:
        continue
    a=a+1
```

```python
# To bypass continue
a = 1 
while a < 5 :
    print(a)
    a = a + 1
    if a == 3:
        continue
    
```
