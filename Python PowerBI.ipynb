### Hotel Data Analysis Report

Create a report to understand the relationship between average daily rate for hotels and the lead time before booking.

    1. You need to compile a collection of excel sheets into a single dataset and filter out cancellations.
    2. Create a Simple linear regression between 2 variables.

```python
# Load the libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress,pearsonr
```

```python
# Load dataset
df1=pd.read_excel(r'C:\Users\praty\OneDrive\Documents\Python+for+Power+BI+Users+-+Resource+Pack\hotel_revenue_historical_full.xlsx',sheet_name='2018')
df2=pd.read_excel(r'C:\Users\praty\OneDrive\Documents\Python+for+Power+BI+Users+-+Resource+Pack\hotel_revenue_historical_full.xlsx',sheet_name='2019')
df3=pd.read_excel(r'C:\Users\praty\OneDrive\Documents\Python+for+Power+BI+Users+-+Resource+Pack\hotel_revenue_historical_full.xlsx',sheet_name='2020')
combine_dataset=pd.concat([df1,df2,df3],ignore_index=True)
```

```python
combine_dataset=pd.concat([df1,df2,df3])
```

```python
combine_dataset.tail(5)
```

```python
combine_dataset.columns
```

```python
# Check the no of cancellations 0 - Not cancel , 1 - Cancelled
combine_dataset['is_canceled'].value_counts()
```

```python
combine_dataset['is_canceled'].value_counts(normalize=True)*100
```

```python
# Filter out cancellations
filtered_data=combine_dataset[combine_dataset['is_canceled']==0]
filtered_data.to_csv('data_without_cancelled_booking.csv')
filtered_data.info()
```

```python
# Univariate and bivariate analysis adr-average daily rate(ADR) wrt to hotel
sns.distplot(filtered_data[filtered_data['hotel']=='City Hotel']['adr'],label='City Hotel')
sns.distplot(filtered_data[filtered_data['hotel']=='Resort Hotel']['adr'],label='Resort Hotel')
resort_mean=filtered_data[filtered_data['hotel']=='Resort Hotel']['adr'].mean()
city_mean=filtered_data[filtered_data['hotel']=='City Hotel']['adr'].mean()
plt.axvline(resort_mean,color='black',linestyle="--",label='Resort Mean')
plt.axvline(city_mean,color='red',linestyle="--",label='City Mean')
plt.title(f"The Overall mean is {filtered_data['adr'].mean():.2f} , Resort_mean is {resort_mean:.2f} and City_mean is {city_mean:.2f}");
plt.legend()
plt.savefig('ADR_distribution')
```

#### From Above chart we observed that Average daily rate of Resort is 90 dollar and City 103 is dollar

```python
sns.distplot(filtered_data['lead_time'])
plt.title(f"The mean is {filtered_data['lead_time'].mean():.2f}");
```

```python
# Plotting distribution Hotel and Lead Time Using Box plot
plt.figure(figsize=(5,5))
sns.boxplot(data=filtered_data,x='hotel',y='lead_time')
resort_mean=filtered_data[filtered_data['hotel']=='Resort Hotel']['lead_time'].mean()
city_mean=filtered_data[filtered_data['hotel']=='City Hotel']['lead_time'].mean()
plt.title(f"The Overall mean is {filtered_data['lead_time'].mean():.2f} , Resort_mean is {resort_mean:.2f} and city_mean is {city_mean:.2f}");
```

#### From Above chart we observed that the Resort Hotels are advance booked around 77 days earler and City Hotels are booked around 76 days earlier

```python
# LinearRegression
linregress(filtered_data['lead_time'],filtered_data['adr'])
```

```python
# LinearRegression
slope=linregress(filtered_data['lead_time'],filtered_data['adr'])[0]
slope
```

```python
intercept=linregress(filtered_data['lead_time'],filtered_data['adr'])[1]
intercept
```

```python
r2=linregress(filtered_data['lead_time'],filtered_data['adr'])[2]*100
r2
```

```python
regression_table=pd.DataFrame({'Name':['slope','intercept','r2'],'Values':[slope,intercept,r2]})
regression_table
```

#### So for every one day in lead time the slope increses 0.008856

```python
# Describe the relationship
from scipy.stats import pearsonr
correlation = pearsonr(filtered_data['adr'],filtered_data['lead_time'])[0]
slope=linregress(filtered_data['adr'],filtered_data['lead_time'])[0]
intercept=linregress(filtered_data['lead_time'],filtered_data['adr'])[0]
regression_table=pd.DataFrame({'Name':['correlation','slope','intercept'],'Coefficients':[correlation,slope,intercept]})
regression_table
```
